#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
df_users = pd.read_csv('users_table.csv')
df_users.head()
```

#### 3. Rename Id column to userId

```python
colum_names = list(df_users.columns)
colum_names
colum_names[0] = 'userId'
df_users.columns = colum_names
df_users.head()

#more straight fordward solution with df.rename, do not forget to use inplace=True to keep otherwise, df = df...
#df_users.rename(columns={'userId': 'id'}, inplace=True)
#df_users.head()
```

#### 4. Import posts table:

```python
df_posts = pd.read_csv('posts_table.csv')
df_posts.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
df_posts.rename(columns = {'Id':'postId', 'OwnerUserId':'userId'}, inplace=True)
df_posts.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_df = df_users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_df = df_posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]

print(users_df.shape)
display(users_df[users_df['userId']==-1])

print(posts_df.shape)
display(posts_df[posts_df['userId']==-1])



```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(left = users_df, right = posts_df, left_on='userId', right_on='userId', how='left') # if used inner we will loose the rows if userId not found in posts_df
merged_df

#I merged so we can't loose any king of data but the way of merging depends on the goal. maybe we are not interested in the users that don't have any posts or whatever
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
merged_df.isnull().sum()
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#I will replace by 0 just because of the way of merging, we have usersIds that dont appear in the postId data, which I assume it is because of no posting.
#I dont want to loose any king of data so I am not cleaning and adding some meaning to the NaNs
#merged_df[merged_df['postId']==0]
columns_to_fill = ['postId', 'Score', 'ViewCount', 'CommentCount']
merged_df[columns_to_fill] = merged_df[columns_to_fill].fillna(0)  #maybe with postId we can put -1 or leave NaN but considering there is no with 0 (checked) we will leave it as 0

display(merged_df.head(50))
display(merged_df.tail(50))
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
merged_df['userId'] = merged_df['userId'].astype(int)
merged_df['postId'] = merged_df['postId'].astype(int)  #take intoaccount that postid0 means no posts
merged_df['ViewCount'] = merged_df['ViewCount'].astype(int)
merged_df['CommentCount'] = merged_df['CommentCount'].astype(int)

display(merged_df.head(50))
display(merged_df.sample(50))

```

```python
#We can get some "outliers"

#Users that have not published any post
users_with_no_posts = merged_df[merged_df['postId']==0]
users_with_no_posts.to_csv('users_with_no_posts.csv')

display(merged_df.describe())
most_popular_post = merged_df[merged_df['ViewCount']>100000]
most_popular_post.to_csv('most_popular_post.csv')

most_commented_post = merged_df[merged_df['CommentCount']==45]
most_commented_post.to_csv('most_commented_post.csv')

#and we can keep adding kpis
```
